## <font color='157699'> **Programación orientada a objetos (POO)**

<b>Objetos predefinidos:</b> en los lenguajes de programación todo elemento de información que usamos es un objeto que tiene un tipo de dato. Por lo general, los objetos que podemos usar vienen determinados por los tipos de datos que permita el lenguaje. Tal es el caso de Python, en donde podemos usar objetos del tipo lista, tupla, entero, <em>string</em>, etc. Para usar estos objetos en nuestros programas definimos instancias; por ejemplo, podemos definir L=[1,2,3,4] para significar que L es una instancia del objeto lista. La representación interna del objeto de tipo lista se materializa mediante la estructura de datos 'lista ligada', que  usa apuntadores para conectar los campos o celdas que la conforman, como se muestra en la figura 1. ilus1.png
<caption><center><b><font color='157699'> Figura 1. Lista ligada.</font></center></caption></b>

Un programador no necesita conocer cómo se representa de forma interna el objeto de tipo lista para poder usarlo en sus programas. Por esto se dice que la representación interna de los objetos es privada o no accesible por terceros. Quien desee definir y manipular en su programa un objeto del tipo lista primero debe definir una instancia de este objeto y luego recurrir a las operaciones o métodos que ese objeto tiene predefinidas. En el caso del objeto de tipo lista se pueden usar las siguientes operaciones:

<center>$L[i:j]$, &emsp; $len(L)$, &emsp; $min(L)$, &emsp;$max(L)$,&emsp; $L.append()$, &emsp;$L.count()$</center>

Con estas operaciones se garantizan dos cosas: que solo se realicen las operaciones autorizadas y que el comportamiento del objeto de tipo lista siempre sea el esperado.

<b>Objetos definidos por el usuario: </b> si queremos habilitar que cada programador cree sus propios objetos, más allá de los que vienen predefinidos, se puede usar POO. En esta, los objetos serán abstracciones de elementos de información ―simples o incluso bastante complejos― que tendrán una representación interna (por lo general, oculta a terceros), y una interfaz para interactuar con ellos mediante métodos (operaciones) los cuales se usarán para definir el comportamiento que puede tener ese objeto. Para definir un nuevo objeto a usar en nuestro programa debemos definir las clases.

## <font color='157699'> **¿Qué es una clase en POO?**

Es el nombre que se le da a un conjunto de objetos tomados del mundo real y se nombran por medio de un sustantivo del lenguaje natural. Es usual que  representen un conjuntos de entidades, conceptos o elementos de información útiles para escribir nuestros programas; por ejemplo, la clase "Tarjeta" representa al conjunto de tarjetas que expide una entidad financiera, la clase "Materia" representa el conjunto de materias que se dictan en una universidad. <br>

La definición de la clase es una especie de plantilla que nos permite especificar la manera cómo se comportan los individuos que pertenecen a ella. Cada individuo creado a partir de esta plantilla se denomina instancia u objeto de la clase. La plantilla incluye tanto un conjunto de atributos, que suelen estar ocultos a terceros, como un conjunto de métodos que regulan el comportamiento de las instancias.<br>

La figura 2 muestra la correspondencia entre los tipos predefinidos en el lenguaje y las clases definidas por el programador. Nota que en ambos casos se instancian objetos bien sea del tipo predefinido en el lenguaje o de las clases definidas por el programador. Tanto los tipos de datos como las clases tienen operaciones o métodos que pueden ser usados para consultar o modificar el comportamiento/estado de las instancias. Para su uso no necesitas conocer los detalles de cómo están implementados.

## <font color='5adcff'> **Definición de una clase en Python**

Supón que necesitamos escribir una clase en Python para definir los datos y las operaciones asociadas a las tarjetas que usan los clientes de un banco. Con el código escrito para la clase debe ser posible crear tarjetas para el banco  "NBANK" y ejecutar operaciones débito, crédito, así como consultar el saldo. Debes tener presente que existen dos tipos adicionales de tarjeta para clientes. La <em>tarjeta VIP</em>, que se diferencia con la tarjeta corriente en que permite a los clientes un sobregiro por un valor autorizado con anterioridad. La <em>tarjeta joven</em>, que se expide solo por un año y por cada consumo mayor de cierta cantidad  en  COP, el  banco te otorga un bono de cierto valor también en  COP; ambas cifras son las mismas para todas la tarjetas de este tipo.

### <font color='5adcff'> **La definición de la clase y su constructor**

<font color='000000'><li>Según el código que se muestra abajo, la clase lleva por nombre Tarjeta, tiene una variable de clase llamada emisor para indicar que todos las instancias que se creen serán tarjetas emitidas por el "NBKAN". </li>
<li>El método constructor de la clase crea instancias o tarjetas individuales para los clientes y para cada tarjeta se debe indicar el número asignado, la clave y el saldo. La tarjeta inicia en estado activa y con fecha de caducidad en dos años. </li>
<li>Nota que el parámetro <b>self</b> aparece en el método <em>__init__</em> para hacer referencia al objeto instanciado de esa clase sobre el cual se está invocando dicho método.</li>
</ul>

```python
from datetime import datetime, timedelta
class Tarjeta:
        """
        Esta clase define los datos y las operaciones necesarias para gestionar tarjetas del banco
        """
        emisor  = "NBank"                                                       # variable de clase.

        def __init__(self, numero, saldo, clave):                               # constructor de objetos de la clase. Se usa para activar cada tarjeta
            self._numero = numero                                               # Atributo número de la tarjeta                - variable de instancia
            self._saldo  = saldo                                                # saldo inicial asignado a la tarjeta.         - variable de instancia
            self._clave  = clave                                                # asignación de clave                          - variable de instancia
            self._estado = "activa"                                             # definición del estado de la tarjeta          - variable de instancia
            self._fecha_caducidad  = datetime.now() + timedelta(days=365*2)     # define fecha de vencimiento                  - variable de instancia                               #
            return
```


El elemento de información 'mi_tarjeta' es una instancia de la clase 'Tarjeta'. En su creación se especifican los parámetros requeridos por el metodo __init__ (self, número, saldo, clave); en este caso el número es 1, el saldo son 1000 COP y la clave es 123.

```python
mi_Tarjeta= Tarjeta(1,1000, '123')
```

### <font color='5adcff'> **La definición de los métodos restantes en la clase**

Los demás métodos de la clase son funciones que solo se ejecutan sobre instancias de esta. Cada método tiene como primer argumento <em>self</em>, para hacer referencia al objeto instanciado de esa clase sobre el cual se está invocando dicho método. Los métodos pueden tener argumentos adicionales y que se comportan como cualquier función; es decir, tienen parámetros, realizan alguna tarea y devuelven uno o varios valores. En el código siguiente se especifican los otros métodos de la clase Tarjeta, tales como debitar, acreditar, mostrar_saldo y anular tarjeta.

```python
from datetime import datetime, timedelta

class Tarjeta:
        """
        Esta clase define los datos y las operaciones necesarias para gestionar tarjetas de crédito
        """
        emisor  = "NBank"                                                       # variable de clase.

        def __init__(self, numero, saldo, clave):                               # constructor de objetos de la clase. Se usa para activar cada tarjeta
            self._numero       = numero                                         # Atributo número de la tarjeta                - variable de instancia
            self._saldo        = saldo                                          # saldo inicial asignado a la tarjeta.          - variable de instancia
            self._clave        = clave                                          # asignación de clave                           - variable de instancia
            self._estado       = "activa"                                       # definición del estado de la tarjeta           - variable de instancia
            self._fecha_caducidad  = datetime.now() + timedelta(days=365*2)     # define fecha de vencimiento                 - variable de instancia                               #
            return

        def debitar(self, cantidad):                                           # método para abonar al saldo de la tarjeta
               if  self._estado == "activa" :
                   self._saldo  += cantidad           # modifico el saldo de esa tarjet
                   return True
               return False

        def acreditar(self, cantidad, clave):                                  # método usado para registrar pagos realizados con la tarjeta.
            if self._estado == "activa"and  self._fecha_caducidad >= datetime.now() and  clave == self._clave and cantidad  <= self._saldo:
               self._saldo  -= cantidad
               return True
            return  False

        def saldo(self):                                                       #  método para consultar el saldo o deuda de la tarjeta
            return  self._saldo

        def anular(self):                                                      # método para cancelar una tarjeta.
            if self._saldo == 0 :
               self._estado = "inactiva"
            return  self._estado
```

# <font color='157699'> **Subclases**

Una subclase no es otra cosa que un conjunto de objetos que son una especialización de la clase a la que pertenecen. Las subclases heredan todos los atributos y métodos de la clase, pero pueden contener atributos y métodos adicionales. Una propiedad bastante importante de las subclases es: pueden sobreescribir métodos definidos en la clase a la que pertenecen.

La subclase Tarjeta_VIP hereda los métodos debitar, saldo y anular de la clase Tarjeta, pero sobreescribe el método acreditar y agrega el método sobregiro. Por su parte, la Tarjeta_Joven  hereda los métodos debitar, saldo y anular de Tarjeta, pero sobreescribe el método acreditar y agrega el método bono.

```python
from datetime import datetime, timedelta
class Tarjeta_Joven(Tarjeta):
        """
        Subclase de la clase Tarjeta.  Es decir, es un tipo especial de Tarjeta.
        Hereda todos los métodos de Tarjeta. Sobreescribe los  método __init()__  y acreditar(); además agrega el método  bono()
        """
        valor_bono = 10
        compra_minima = 100
        def __init__(self, numero, saldo, clave ):                              # sobreescribe el método __init__ de la superclase
            self._numero          = numero                                      # Atributo número de la tarjeta               - variable de instancia
            self._saldo           = saldo                                       # Cupo asignado a la tarjeta.                 - variable de instancia
            self._clave           = clave                                       # creación del atributo clave                 - variable de instancia
            self._estado          = "activa"
            self._fecha_caducidad = datetime.now() + timedelta(days=365)       # define fecha de vencimiento                  - variable de instancia
            self._bono_actual     = 0


        def bono(self):                                                        # Método exclusivo de la clase Tarjeta_VIP
            return print('Bono acumulado', self._bono_actual, 'COP')

        def acreditar(self, cantidad, clave):                                    # método usado para registrar pagos realizados con la tarjeta.
            if self._estado == "activa" and self._fecha_caducidad >= datetime.now() and clave == self._clave and cantidad <= self._saldo:
              if cantidad >= Tarjeta_Joven.compra_minima:
                  self._saldo -= cantidad - self._bono_actual
                  self._bono_actual += Tarjeta_Joven.valor_bono
              return True
            return False
```

## <font color='5adcff'> **Bondades de la POO**

Por lo general, los lenguajes usados para la programación orientada a objetos habilitan el uso de las siguientes características.
### Abstracción

Permite representar el mundo a programar como una serie de clases, cada una con estados y un comportamiento definidos por el programador que las crea. Como ejemplo podemos tener la clase Tarjeta y las subclases Tarjeta_VIP, Tarjeta_Platinum, Tarjeta_Joven.

### Encapsulamiento

Posibilita ocultar la implementación de la clase a otros programadores; es decir, estos no requieren conocer el código de los métodos para usarlos. Esta propiedad también permite establecer restricciones de acceso a los métodos y atributos de la clase. La visibilidad de los elementos  de la clase se define mediante los siguientes especificadores.

<b>Public</b>. Los métodos y variables son accesibles desde otras clases.</li>

<b>Private</b>. Si se definieron los métodos o variables como <b>private</b> no son accesibles desde el exterior de la clase, solo pueden ser usados desde la clase que los define. Se especifican mediante el doble guion bajo, como el caso del método constructor __init(  )__.</li>

<b>Protected</b>. Los métodos y variables definidas de esta manera en una clase solo son accesibles en clases derivadas de esta. Se definen usando un solo guion bajo. </li>
</ul>

<font color='157699'>Nota: <font color='000000'> a diferencia de otros lenguajes OO como Java y C++, Python no valida la visibilidad de los elementos de la clase. En la sección 7 volveremos sobre este asunto.


### Herencia

Nos permite definir múltiples subclases a partir de una clase ya definida, lo que habilita la reutilización de código ya escrito. Las subclases heredan las variables y los métodos de la clase. En el ejemplo, la clase Tarjeta_VIP hereda los métodos de la clase Tarjeta.

### Polimorfismo

Nos posibilita modificar los métodos y atributos de una clase en las subclases que defina. En el ejemplo, la clase Tarjeta_VIP sobreescribe los métodos <b>__init( )__ </b> y <b>acreditar()</b> de la clase Tarjeta.
