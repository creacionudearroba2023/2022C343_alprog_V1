## <font color='3E779E'> **Qu es la hoja de estilo PEP 8?**</font>



Python Enhancement Proposal 8 (PEP 8) es una gua de estilo para escribir cdigo Python, se utiliza de forma amplia en la comunidad de programacin de Python. Esta gua proporciona recomendaciones para la escritura de cdigo de manera consistente y legible, su fin es mejorar su calidad, mantenibilidad y legibilidad.
La hoja de estilo PEP define conceptos como la indentacin, la longitud de las lneas, las lneas en blanco y la gua de importacin de otros paquetes.
En general, algunos puntos clave son:
* Utiliza sangra de cuatro espacios para indicar bloques de cdigo. No se recomienda usar tabulaciones.
* Limita la longitud de la lnea a 79 caracteres. Si debe ser ms larga, se puede dividir en varias lneas con la convencin de continuacin de lnea (\).
* Usa nombres de variables y funciones que sean claros y descriptivos. Utiliza maysculas y minsculas de forma coherente para mejorar la legibilidad.
* Utiliza espacios alrededor de operadores y despus de las comas para mejorar la legibilidad.
* Escribe cadenas de documentacin para todas las funciones, clases y mdulos.
* Importa mdulos en lneas separadas y ordnalos de manera alfabtica.
* No uses espacios en blanco al final de la lnea.
* Usa los nombres de clase **CamelCase**, los nombres de funcin en minsculas con guiones bajos y los nombres de constantes en maysculas.

## <font color='45adcff'> **CamelCase** </font>
En Python, **CamelCase** es un mtodo de codificacin para nombrar variables, funciones y clases. Se utiliza para crear identificadores de varias palabras, donde cada una (excepto la primera) comienza con una letra mayscula. El propsito de las convenciones de CamelCase es mejorar la legibilidad y comprensin del cdigo. Hay dos estilos:

**1. UpperCamelCase o PascalCase**: en este estilo la primera letra de cada palabra se escribe en mayscula, incluso la primera palabra, por ejemplo:

```python
class MiClase:
    def miMetodo(self):
        pass
```

**2. lowerCamelCase**: en este estilo la primera letra de cada palabra, excepto la primera palabra, se escribe en mayscula. Por ejemplo:

```python
miVariable = 5
def miFuncion():
    pass
```

## <font color='45adcff'> **Indentacin** </font>
En Python, a diferencia de otros lenguajes de programacin que usan llaves o palabras clave para separar bloques de cdigo, se acude a la **indentacin** para definir la estructura y la jerarqua del cdigo. La indentacin tambien se puede definir como una sangra o grupo de espacios que se muestran al comienzo de una lnea. Esta se utiliza para agrupar sentencias que forman parte de una estructura de control, una funcin, un bucle o una sentencia condicional. Todas las declaraciones con la misma indentacin se consideran parte del mismo bloque de cdigo.

Python no usa {} o *begin - end* para designar bloques de cdigo como otros lenguajes de programacin, sino que utiliza bloques indentados para indicar que un determinado bloque de cdigo pertenece a ese espacio. Python solo aplica la cantidad de indentacin para definir la lgica.

## <font color='45adcff'> **Identificadores** </font>
En Python, es un nombre que se usa para identificar una variable, funcin, clase, mdulo u otro objeto en el cdigo. Estos nombres sirven para referirse y realizar operaciones en objetos; adems, seguir estas reglas y convenciones permite mantener un cdigo legible y consistente. Los principales elementos, reglas o conveciones que debes tener en cuenta para definir los identificadores son:

**Reglas bsicas**
* Los identificadores pueden contener letras (maysculas o minsculas), nmeros y guiones bajos. Por ejemplo: comienzan con una letra de la 'A' a la 'Z' (minscula o mayscula) o un guion bajo, seguido de cero o ms letras, guiones bajos y dgitos (del 0 al 9).

## <font color='45adcff'> **Qu es *name mangling*?** </font>
En Python, es una tcnica utilizada para cambiar el nombre de los atributos de una clase al agregar un prefijo especial. Su objetivo es evitar colisiones de nombres en clases que heredan propiedades de clases principales o padre.

Cuando se agrega un guion bajo doble (__) al nombre de un atributo de clase se aplica de manera automtica el *name mangling*. Cambia el nombre original del atributo al adicionar _ClassName antes del nombre original. Esto hace que la propiedad est "oculta" y solo est disponible dentro de la propia clase y sus subclases, de manera que evita colisiones de nombres en la herencia.

La **herencia** es un concepto de programacin orientada a objetos que le permite a una clase heredar atributos y mtodos de otra clase.


```python
class MiClase:
    def __init__(self):
        self.__attribute_private = 2023

    def __metod_private(self):
        print("Es un mtodo privado")

```

## <font color='45adcff'> **Convenciones al nombrar elementos** </font>
Cuando escribes cdigo, todo tiene un nombre: variable, clase, funcin, paquete, mdulo, etc. Por eso es importante hacer que el cdigo sea lo ms legible posible de acuerdo con ciertas reglas. Las clases no tienen los mismos nombres que las funciones y no se recomienda usar nombres como *a* o *x*, ya que transmiten menos informacin. A continuacin, te damos algunas recomendaciones:
* Evita usar palabras reservadas. Por ejemplo, si es necesario usar 'class', acude a 'class_' como alternativa (ver sesin de identificadores).
* Evita utilizar las letra *l* (letra L)  o/y *I* (I latina), ya que se pueden confundir.
* Usa el guion bajo (_) ms el nombre de la variable (ej:  _variable) para especificar uso interno. Por ejemplo: 'from m import'; as no importara lo que empieza con _.
* Puedes usar '__variable' para invocar el *name mangling* y hacer privadas determinadas variables o mtodos. (Ver sesin sobre *name mangling*).
* Para utilizar elementos como '__init__', que es un mtodo especial y predeterminado para inicializar objetos cuando se crea una instancia de una clase, y otros similares, es necesario recurrir al nombre asignado por Python para su uso y no aplicar alias o nombres alternos.


Python define cmo nombrar cada tipo tipo de elemento de su lenguaje. Te mostramos algunas buenas prcticas:

* **Funciones**: para nombrarlas utiliza letras en minscula separadas por  guion bajo: funcion, mi_funcion_de_prueba.
* **Variables**: igual que las funciones: variable, mi_variable.
* **Clases**: usa CamelCase, mayscula y sin barra baja: MiClase, ClaseDePrueba.
* **Mtodos**: al igual que en las funciones, usa *snake case*: metodo, mi_metodo.
Constantes: nmbralas con maysculas y sepralas con barra baja: UNA_CONSTANTE, OTRA_CONSTANTE.
* **Mdulos**: igual que las funciones: modulo.py, mi_modulo.py.
* **Paquetes**: en minsculas, pero sin separar por barra bajas: paquete, mipaquete.


```python
#Ejemplo de como nombrar
class MiClase():
    def mi_primer_metodo(self, variable_a, variable_b):
        return (variable_a + variable_b)


mi_objeto = MiClase()
print(mi_objeto.mi_primer_metodo(5, 5))
```

Los **linters** son programas incoporados en Python que realizan un anlisis esttico del cdigo, permiten corregir errores al momento de la escritura.
A veces puede ser difcil recordar todas las lneas de PEP8, por lo que existen herramientas que pueden autocorregir o marcar reas problemticas en el cdigo.
* Hay dos tipos: los linters como Flake8 o pycodestyle.
* Autoformatters: se limitan a indicarnos dnde el cdigo no cumple con las normas. En ciertos casos realiza las correcciones automticamente.


```python
#Ejemplo Autoformatters
def MiFuncionSuma(A, B, C, imprime = True):
    resultado=A+B+C
    if imprime != False:
        print(resultado)
    return resultado
a          = 4
variable_b = 5
var_c      = 1
```

El comando anterior te permitir recibir una notificacin de cualquier regla que el cdigo est violando, para que puedas corregirla. Es importante tener en cuenta que algunas rutas se autocorregen y otras no. Autopep8 puede corregir de forma automtica cualquier cosa que implique el uso de espacios o lneas en blanco, sin embargo, no cambia los nombres de las variables. Por lo que si rompes las convenciones de nomenclatura de variables, tendrs que corregir de manera manual.
El cdigo anterior viola la siguiente regla:
* E251: uso incorrecto de espacios en imprime = True, debera ser imprime=True.
* E225: los operadores como el + deben usar espacios, A + B + C.
* 712: usar if imprime en vez de if imprime != False.
* E305: despus de la declaracin de una funcin debemos dejar dos espacios en blanco.
* E221: no debemos usar tantos espacios al usar el operador = creando variables.
Tambin tenemos otros problemas relacionados con cmo nombrar a funciones y variables. Las funciones y variables deben ir en *snake case*. Lo veremos en detalle ms adelante.
Al corregirlo quedara de la siguiente forma:
