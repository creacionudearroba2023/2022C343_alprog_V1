**<h2>Ciencias sociales y humanas</h2>**

<p>
Eres un investigador en ciencias humanas y estás interesado en analizar cómo ha evolucionado la popularidad de diferentes géneros literarios (por ejemplo: poesía, novela, ensayo, drama) a lo largo de los últimos 100 años en una región determinada.
</p>
<p>
Para ello, has recopilado datos de ventas y publicaciones de libros, críticas literarias, menciones en medios de comunicación, entre otros, divididos por décadas.
</p>

<p>Realiza las siguientes tareas:</p>
<ol>
<li>Lee el archivo llamado <em><a href="https://udearroba.udea.edu.co/imagescourses/2022C343_alprog_V1/u4/dataset_popularidad_generos_literarios.csv" target="_blank">dataset_popularidad_generos_literarios.csv</a></em>.</li>
<li>Haciendo uso de la librería Numpy, calcula estadísticas básicas sobre los datos: promedios de publicaciones por género, década con más publicaciones totales, etc.</li>
<li>
Organiza los datos en un DataFrame de Pandas y realiza operaciones como filtrado y agrupación para entender mejor los datos. Por ejemplo, podrías querer filtrar solo las décadas donde las novelas superaron a la poesía en publicaciones.
</li>
<li>Utiliza Matplotlib o Seaborn para visualizar la evolución de la popularidad de cada género a lo largo del tiempo, comparar géneros entre sí, etc.</li>

</ol>

</p>


<p>Realiza las siguientes tareas:</p>

</ol>


**<h2> Ciencias de la salud</h2>**

<p>
Eres un investigador en nutrición y estás conduciendo un estudio sobre una nueva dieta diseñada para pacientes con diabetes tipo 2. A lo largo de un año, registraste mensualmente los niveles de glucosa en sangre, peso y otros indicadores, en un grupo de pacientes que siguieron la dieta y un grupo control que mantuvo su alimentación habitual.
</p>

<p>Realiza las siguientes tareas:</p>
<ol>
<li>Lee el arhivo que contiene los niveles promedio mensuales de glucosa en sangre para ambos grupos, este se llama <em><a href="https://udearroba.udea.edu.co/imagescourses/2022C343_alprog_V1/u4/datos_dietas_diabetes.csv" target="_blank">datos_dietas_diabetes.csv</a></em>.</li>
<li>Haciendo uso de la librería Numpy, calcula estadísticas descriptivas como la media, mediana y desviación estándar de los niveles de glucosa para ambos grupos.</li>
<li>
Organiza los datos en un DataFrame de Pandas y compara las variaciones de los niveles de glucosa mes a mes para cada grupo.
</li>
<li>Utiliza Matplotlib o Seaborn para  representar gráficamente la evolución de los niveles de glucosa de ambos grupos a lo largo del año.</li>

</ol>



**<h2> Ciencias de la educación</h2>**

<ol>

Organiza los datos en un DataFrame de Pandas. Filtra y agrupa los datos para analizar el rendimiento según diferentes variables (por ejemplo, año, grado escolar, género).

<li>Utiliza Matplotlib o Seaborn para  graficar la evolución de las calificaciones promedio antes y después de la transición a la educación virtual.</li>
</ol>

</ol>
